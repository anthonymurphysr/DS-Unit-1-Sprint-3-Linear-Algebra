## 1.1 Create a two-dimensional vector and plot it on a graph

```
# Import packages
import numpy as np
import matplotlib.pyplot as plt

# Assign vector
vec = [1, 3]

# Plot vector
plt.arrow(0,0, *vec, head_width=.1, head_length=0.2, color ='black')

# Clean graph with limits and title
# Set x-limits
plt.xlim(0, 1.5)
# Set y-limits
plt.ylim(0, 4)
# Set Title and squash output with ;
plt.title('Two-dimensional vector');
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
# Import Axes3D from mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D

# Create vector array for 3D Plot
vector = np.array([0, 0, 0, .3, .5, .8])

# Zip Vector
X, Y, Z, U, V, W = zip(vector)

# Set fig
fig = plt.figure()

# Set ax to 3D Projection
ax = fig.add_subplot(projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)

#Set x-limit
ax.set_xlim([0, 1])
# Set y-limit
ax.set_ylim([0, 1])
# Set z-limit
ax.set_zlim([0, 1])
# Set title and squash output with ;
plt.title('Three-Dimensional Vector');

# Plot vector
plt.show()
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
# Import norm from LA in numpy
from numpy import linalg as LA

# Assign g
g = np.array([[1],
              [1],
              [1],
              [8]
              ])
# Assign h
h = np.array([[3],
              [3],
              [3],
              [8]
              ])

# Print ||g|| and ||h||
print('||g|| = ', LA.norm(g))
print('||h|| = ', LA.norm(h))
```

---
|| *h* || is longer than || *g* ||.

```
# Set Matrices
a = np.array([[1, 2],
              [3, 4],
              [5, 6]])

b = np.array([2, 4, 6])

c = np.array([[9, 6, 3],
              [4, 7, 11]])

d = np.array([[1, 0, 0], 
              [0, 1, 0],
              [0, 0, 1]])

e = np.array([[1, 3], 
              [5, 7]])
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# Call .dot from numpy 
ae = np.dot(a, e)
cd = np.dot(c, d)
ba = np.dot(b, a)

# Print results of dot notation
print('The product of AE is: ', ae)
print('The product of CD is: ', cd)
print('The product of BA is: ', ba)
```

---
The resulting matrices have the following dimensions:
* AE : [ 3x2 ]
* CD : [ 2x3 ]
* BA : [ 1x2 ]

---
The relation is that the product of the matrices takes on the equivalent row quantity from the first matrix passed.
